```python
import geopandas as gpd
import pydeck as pdk
import json
```

```python
with open("P85EH72.geojson", "r") as f:
    data = json.load(f)
data
```

```python
initial_view_state=pdk.ViewState(
            latitude  = 51.6693206,
            longitude = -9.002953,
            zoom=12,
            pitch=0,
        )
```

```python
geojson = pdk.Layer(
    'GeoJsonLayer',
    data=data,
    opacity=0.8,
    stroked=False,
    filled=True,
    extruded=True,
    wireframe=True,
    get_elevation='properties.name',
    get_fill_color='[255, 255, properties.growth * name]',
    get_line_color=[255, 255, 255],
    pickable=True
)
```

```python
r = pdk.Deck(
    layers=[geojson],
    initial_view_state=initial_view_state,
    map_provider="mapbox",
    map_style=pdk.map_styles.SATELLITE,
    api_keys  = { "mapbox": "pk.eyJ1IjoiZ3JvbDIwMjAiLCJhIjoiY2tvajI0MW5pMDMxMDJ3bzdpN3dzbHBidyJ9.jxYSuCXT1u1eFKMMWwSFVg"}
)

r.to_html()
```
